```python
#imports necesarry packages
import pandas as pd
import json
import os
import re
```

```python
#function for cleaning .json files and converting to a pandas dataframe
def clean_json(file_txt, length):
    small_places_file = file_txt[:length]
    
    #finds all dictionary strings entries and gets the key values for columns
    dicts = re.findall('\{.+\}', small_places_file)
    column_vals = list(eval(dicts[0]).keys())
    
    #evaluates all dictionary strings to dictionary objects and create a data frame
    dicts = list(map(lambda x: list(eval(x).values()), dicts))
    clean_data = pd.DataFrame(dicts, columns=column_vals)
    return clean_data
```

```python
#reads in sample data and displays it
sample_df = pd.read_json("datasets/sample.reviews.json")
sample_df.head()
```

```python
#convert .json place file into pandas dataframe and saves it
places_file = open("datasets/places.clean.json").read()
length = len(places_file)
places_df = clean_json(places_file, length)
places_df.to_csv('places_df.csv')
```

```python
#convert .json reviews file into pandas dataframe and saves it
reviews_file = open("datasets/reviews.clean.json").read()
length = len(reviews_file)
reviews_df = clean_json(reviews_file, length)
reviews_df.to_csv('reviews_df.csv')
```

```python
#convert .json user file into pandas dataframe and saves it
users_file = open("datasets/users.clean.json").read()
length = len(users_file)
users_df = clean_json(users_file, length)
users_df.to_csv('users_df.csv')
```

```python
#read in cleaned places file
places_df = pd.read_csv('places_df.csv')
```

```python
#read in cleaned reviews file
reviews_df = pd.read_csv('reviews_df.csv', lineterminator='\n')
```

```python
#read in cleaned users file
users_df = pd.read_csv('users_df.csv', lineterminator='\n')
```

```python
#drop unecessary indexing row
places_df = places_df.drop(places_df.columns[0], axis=1)
reviews_df = reviews_df.drop(reviews_df.columns[0], axis=1)
users_df = users_df.drop(users_df.columns[0], axis=1)
```

```python
display(places_df.head())
print('places_df rows: ' + str(len(places_df)))
display(reviews_df.head())
print('reviews_df rows: ' + str(len(reviews_df)))
display(users_df.head())
print('user_df rows: ' + str(len(users_df)))
```

```python
#merges user and review dataframes
user_rev = users_df.merge(reviews_df, how='left', validate='1:m')
```

```python
#display merged table
user_rev.head().T
```

```python
#save .csv file
user_rev.to_csv('user_rev.csv')
```

```python
#merges reviews and place dataframes
place_rev = reviews_df.merge(places_df, how='right', validate='m:1')
```

```python
#display merged table
place_rev.head().T
```

```python
#drop data where the gps location is null
place_rev = place_rev[place_rev['gps'].notnull()] 
```

```python
#clean gps data
place_rev['lat'] = place_rev['gps'].apply(lambda x: float(x[1:-1].split(', ')[0]))
place_rev['long'] = place_rev['gps'].apply(lambda x: float(x[1:-1].split(', ')[1]))
```

```python
#drop gps column
place_rev = place_rev.drop('gps', axis=1)
```

```python
#save .csv file
place_rev.to_csv('place_rev.csv')
```
